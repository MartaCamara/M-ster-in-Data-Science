### Ejercicio 2

tenemos que buscar el top 10 de aeropuertos por número de llegadas

```python
# vamos a leer sólo las primeras 100.000 líneas

b = pd.read_csv('bookings.csv.bz2',sep='^',nrows=1e5)
```

```python
b.head()
```

```python
# agrupamos por la columna arr_port creando un dataframe con groupby de la columna aeropuerto

df_aeropuerto = b.groupby("arr_port")
```

```python
# aplicamos la función suma a todos los campos del dataframe groupby

df_sumaeropuerto = df_aeropuerto.aggregate(sum)

# otra forma sería  df_aeropuerto.pax.sum()
```

```python
# ordenar de mayor a menos por la columna pax

df_sumaeropuerto.sort("pax", ascending=False)
```

```python
# seleccionamos sólo los 10 primeros

top_ten_aer = df_sumaeropuerto.sort("pax", ascending=False).head(10)
```

```python
top_ten_aer["pax"]
```

```python
# Solución propuesta

b = pd.read_csv('bookings.csv.bz2', sep='^')
g = b.groupby('arr_port')
s = g.pax.sum()
d = s.reset_index()
d.sort('pax', ascending=False).head(10)
```
